**2.BIVARIATE AND MULTIVARIATE ANALYSIS**

```python
tips= sns.load_dataset('tips')
```

here "tips" is a built-in dataset of the seaborn library

```python
tips.head()
```

```python
titanic=pd.read_csv('/content/train.csv')
```

```python
titanic.sample(5)
```

```python
flights= sns.load_dataset('flights')
flights.head(3)
```

```python
iris= sns.load_dataset('iris')
iris.head()
```

1.Scatterplot (numerical-numerical).     (to remember:- hue, style, size)

```python
sns.scatterplot(x="total_bill", y="tip", data=tips) #each dot represents a customer
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex") #this is a multivariate analysis ans inside hue we have given the third column and each dot represents a customer
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="smoker")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="smoker", size="size")
```

2.Bar Plot (numerical-categorical)

```python
sns.barplot(x="Pclass", y="Age", data=titanic)
```

```python
sns.barplot(x="Pclass", y="Age", hue="Sex", data=titanic)
```

3.Boxplot (numerical-categorical)

```python
sns.boxenplot(x="Sex",y="Age", data=titanic)
```

```python
sns.boxplot(x="Sex",y="Age", data=titanic)
```

```python
sns.boxplot(x="Sex",y="Age", data=titanic, hue="Survived")
```

4.Distplot(numerical- categorical). use "kdeplot"

```python
import seaborn as sns

sns.kdeplot(data=df[df['Survived']==0], x="Age", label="Did not survive")
sns.kdeplot(data=df[df['Survived']==1], x="Age", label="Survived")
```

5.Heatmap (categorical-categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived']) #Row = Passenger class (1, 2, 3)

                                                   #Column = Survival status (0 = Died, 1 = Survived)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
flights.pivot_table(values='passengers',index='month',columns='year') #this is so confusion so use heatmap
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
titanic.groupby('Pclass')['Survived'].mean()*100 #it is telling the mean survival percentage of each p class
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')
```

6.Cluster map(categorical- categorical)

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

7.Pairplot (collects all the numerical cols and plots the scatter plot for all)

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue="species")
```

8.Lineplot (numerical- numerical)

```python
new=flights.groupby('year')['passengers'].sum().reset_index() #This makes it explicit that you only want the sum of passengers
```

```python
sns.lineplot(x=new['year'], y=new['passengers'])
```
